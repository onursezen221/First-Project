# U.S. Medical Insurance Costs

```python
import csv 
with open("insurance.csv") as insurance:
    insurance_dict = csv.DictReader(insurance)
    ages= list()
    sex = list()
    bmi = []
    children = []
    smoker = []
    region = []
    charges = []
    for i in insurance_dict:
        ages.append(i["age"])
        sex.append(i["sex"])
        bmi.append(i["bmi"])
        children.append(i["children"])
        smoker.append(i["smoker"])
        region.append(i["region"])
        charges.append(i["charges"])
class Citizen:
    def __init__(self, age, sex, bmi, children, smoker, region, charges):
        self.age = int(age)
        self.sex = sex
        self.bmi = float(bmi)
        self.children = int(children)
        self.smoker = smoker 
        self.region = region
        self.charges = float(charges)
    def agesituation(self):
        if 18 <= self.age < 25:
            return "Very Young"
        elif 25 <= self.age < 32:
            return "Young"
        elif 32 <= self.age < 40:
            return "Middle Age"
        elif 40 <= self.age < 50:
            return "Slightly Old"
        elif 50 <= self.age < 56:
            return "Old"
        else:
            return "Quite Old"
    def bmisituation(self):
        if self.bmi < 18.5:
            return "Underweight"
        elif self.bmi < 25:
            return "Normal"
        elif self.bmi < 30:
            return "Overweight"
        else:
            return "Obese"
    def smokerness(self):
        if self.smoker == "yes":
            return True
        else:
            return False
    def maleness(self):
        if self.sex == "male":
            return True
        else:
            return False

citizens = []
for i in ages: 
    idx = ages.index(i)
    citizens.append(Citizen(i, sex[idx], bmi[idx], children[idx], smoker[idx], region[idx], charges[idx]))

def average_age(citizens):
    total = 0
    for i in citizens:
        total += i.age
    return total / len(citizens)

def average_bmi(citizens):
    total = 0
    for i in citizens:
        total += i.bmi
    return total / len(citizens)

def average_children(citizens):
    total = 0
    for i in citizens:
        total += i.children
    return total / len(citizens)

def age_analysis(citizens):
    veryyoung, young, middleage, slightlyold, old, veryold = 0,0,0,0,0,0
    bmi_list, bmi_list1, bmi_list2, bmi_list3, bmi_list4, bmi_list5 = [],[],[],[],[],[]
    smoker, smoker1, smoker2, smoker3, smoker4, smoker5 = 0,0,0,0,0,0
    for i in citizens:
        if i.agesituation() == "Very Young":
            veryyoung += 1
            bmi_list.append(i.bmi)
            if i.smokerness():
                smoker+=1
        elif i.agesituation() == "Young":
            young += 1
            bmi_list1.append(i.bmi)
            if i.smokerness():
                smoker1+=1
        elif i.agesituation() == "Middle Age":
            middleage += 1
            bmi_list2.append(i.bmi)
            if i.smokerness():
                smoker2+=1
        elif i.agesituation() == "Slightly Old":
            slightlyold += 1
            bmi_list3.append(i.bmi)
            if i.smokerness():
                smoker3+=1
        elif i.agesituation() == "Old":
            old += 1
            bmi_list4.append(i.bmi)
            if i.smokerness():
                smoker4+=1
        else:
            veryold += 1
            bmi_list5.append(i.bmi)
            if i.smokerness():
                smoker5+=1
    veryyoungbmi = sum(bmi_list) / veryyoung
    youngbmi = sum(bmi_list1) / young 
    middleagebmi = sum(bmi_list2) / middleage
    slightlyoldbmi = sum(bmi_list3) / slightlyold 
    oldbmi = sum(bmi_list4) / old 
    veryoldbmi = sum(bmi_list5) / veryold 
    
    veryyoungsmoker = smoker / veryyoung
    youngsmoker = smoker1 / young 
    middleagesmoker = smoker2 / middleage
    slightlyoldsmoker = smoker3 / slightlyold 
    oldsmoker = smoker4 / old 
    veryoldsmoker = smoker5 / veryold    
    print("Very Young BMI Average: {}\nYoung BMI Average: {}\nMiddle Age BMI Average: {}\nSlightly Old BMI Average: {}\nOld BMI Average: {}\nVery Old BMI Average: {}".format(veryyoungbmi,youngbmi,middleagebmi, slightlyoldbmi, oldbmi, veryoldbmi)) 
    print("Very Young Smoker %:", veryyoungsmoker,"\nYoung Smoker %:", youngsmoker, "\nMiddle Age Smoker %:", middleagesmoker, "\nSlightly Old Smoker %:", slightlyoldsmoker, "\nOld Smoker %:", oldsmoker, "\nVery Old Smoker %:", veryoldsmoker)
    return [veryyoungbmi, youngbmi, middleagebmi, slightlyoldbmi, oldbmi, veryoldbmi], [veryyoungsmoker, youngsmoker, middleagesmoker, slightlyoldsmoker, oldsmoker, veryoldsmoker]
print(age_analysis(citizens))

    

```
